```python
#for a_row_im in all_row_images
def plaster_paintings(some_X,corner_i=None):
    if not corner_i:
        corner_i = int(np.sqrt(len(some_X)))
    full_matrices = np.hstack(some_X[:corner_i])
    for row_num in xrange(1,corner_i):
        this_row = np.hstack((some_X[row_num*corner_i:(row_num+1)*corner_i]))
        full_matrices = np.vstack((full_matrices,this_row))
    return full_matrices
```

```python
y = pd.DataFrame(y)
y.columns = ['Abstract','Cubist','Impressionist','Realist']
```

```python
idx_r = y[y['Realist'] == 1].index.tolist()[:1225]
idx_i = y[y['Impressionist'] == 1].index.tolist()[:1225]
idx_c = y[y['Cubist'] == 1].index.tolist()[:1225]
idx_a = y[y['Abstract'] == 1].index.tolist()[:1225]

X_r = np.squeeze(X[idx_r])
X_i = np.squeeze(X[idx_i])
X_c = np.squeeze(X[idx_c])
X_a = np.squeeze(X[idx_a])
all_X = [X_r,X_i,X_c,X_a]
```

```python
pasted_X = []
for each_X in all_X:
    pasted_X.append(plaster_paintings(each_X,20))
```
